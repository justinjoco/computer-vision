# Testing MNIST dataset using Sci-Kit Classifiers

- KNN (N=1,3,5)
- Multi-layer perceptron with one and two hidden layers
- Support Vector Machines (Linear, RBF, Polynomial)

```python
import time
import io
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.arff import loadarff

from sklearn.datasets import get_data_home
from sklearn.externals.joblib import Memory
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils import check_random_state
from urllib.request import urlopen

import warnings
warnings.filterwarnings('ignore')
```

```python
memory = Memory(get_data_home())
@memory.cache()
def fetch_mnist():
    content = urlopen(
        'https://www.openml.org/data/download/52667/mnist_784.arff').read()
    data, meta = loadarff(io.StringIO(content.decode('utf8')))
    data = data.view([('pixels', '<f8', 784), ('class', '|S1')])
    return data['pixels'], data['class']
X, y = fetch_mnist()
```

```python
X = X / 255.

###### NEW  Refromat the the labels to be integers rather than byte arrays
y_trans = []
for i in range(len(y)):
    y_trans.append(int(y[i]))
y = np.asarray(y_trans)

train_size = 30000
test_size = 500
print("Training size = {}; Test size = {}".format(train_size, test_size))

X_train, X_test = X[:train_size], X[train_size:train_size+test_size]
y_train, y_test = y[:train_size], y[train_size:train_size+test_size]
```

```python
### Bounus 1: show the first 40 images
### ALways a good idea to validate that the data appears as you expect
### 
### for sci-kit learn the images are represented as vectors of 784 elements
### currently scaled from 0 to 1

for i in range(50):
    l1_plot = plt.subplot(5, 10, i + 1)
    l1_plot.imshow(255 * X_test[i].reshape(28, 28), interpolation='nearest',
                   cmap=plt.cm.gray)
    l1_plot.set_xticks(())
    l1_plot.set_yticks(())
    #l1_plot.set_xlabel('Class %s' % y_test[i].decode())
    l1_plot.set_xlabel('%i' % int(y_test[i]))
plt.suptitle('Test image Examples')
plt.show()
```

```python
## Data standardization
## by the mean and standared deviation of the training set
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
start = time.time()
knn.fit(X_train, y_train)
end = time.time()

print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))
start = time.time()
score = knn.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score 1-NN: %.4f" % score)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
start = time.time()
knn.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = knn.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score 3-NN: %.4f" % score)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
start = time.time()
knn.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = knn.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score 5-NN: %.4f" % score)
```

## KNN Discussion

Based on the above results, despite having a much larger training size than testing size, kNN took longer to predict test values than train on the training data. As a result, shrinking the training size from 60k to 30k did not have much of an impact on training time. However, decreasing the testing size from 10k to 500 significantly decreased the test time.

Overall, I observed that 1NN, 3NN, and 5NN made very good predictions on the input data with at least 94.2% correctness, with very small accuracy discrepancies between them.

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,))
start = time.time()
mlp.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = mlp.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score MLP, 1 hidden layer: %.4f" % score)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,10))
# Train the classifier
start = time.time()
mlp.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = mlp.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score MLP, 2 hidden layers: %.4f" % score)
```

## Multi-Layer Perceptron Discussion

Based on the above results and theory on deep learning, training time took proportionally much more time than testing time due to the feedforward and backpropogation operations in the training stage. Since these operations learn the weight and bias vectors, the testing stage is mostly about operating the new input on these learned vectors, which takes relatively no time at all.

I note that, as one might imagine, adding a second hidden layer increased training time. As the number of hidden layers increase, along with the number of nodes per layer, the overall complexity of the neural network will increase, heightening training time.

As a result, decreasing the training set had a signficant improvement on training time; on the other hand, decreasing the test size did not seem to have a very meaningful change on test time.

I observe that although both neural nets have an accuracy of about 90%, the 2 hidden layer NN had slightly worse accuracy, most likely due to overfitting on the training set.

```python
svm = SVC(kernel="linear")
start = time.time()
svm.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = svm.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score SVM, linear %.4f" % score)
```

```python
svm = SVC(kernel="rbf")
start = time.time()
svm.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = svm.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score SVM, radial basis function %.4f" % score)
```

```python
svm = SVC(kernel="poly")
start = time.time()
svm.fit(X_train, y_train)
end = time.time()
print("Training time elapsed: {:.4f} minutes".format((end-start)/60.0))

start = time.time()
score = svm.score(X_test, y_test)
end = time.time()
print("Test time elapsed: {:.4f} minutes".format((end-start)/60.0))
print("Test score SVM, polynomial %.4f" % score)
```

## SVM Discussion

From the above, I see that the SVM with the rbf kernel had the best accuracy at a whopping 97.6%. Though its training and test time were greater than that of the linear SVM, these were still less than that of the poly-SVM. Though poly-SVM had the second highest accuracy, its training and test time were noticably longer than that of any of the others.

Like with the MLP's, decreasing the training set had a signficant improvement on training time(moreso than the MLP's, certainly). Additionally, decreasing the test size did noticeably decrease testing time.

If I were to pick one of these kernels in an SVM for an image processing application, I would choose rbf due to its high accuracy, yet very respectable training and test time.
